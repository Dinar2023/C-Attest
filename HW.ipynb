```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

```python
list_human=[]
list_robot=[]
for item in data['whoAmI']:
  if item == 'human':
    list_human.append(1)
    list_robot.append(0)
  elif item == 'robot':
    list_human.append(0)
    list_robot.append(1)
df = pd.DataFrame({'human': list_human, 'robot': list_robot})
df.head(20)
```
